```python
# Here i will create a new feature Age, that will done out of Year_Birth, as age typically gives better information in this instance. In the same light, I will extract the date in the Dt_Customer 
# feature and then subtract the date that each customer registered then take it all the way down to the year as recent as we have the newest joining customer, as indicated in the data. I will create a feature 
# Join_duration from this.
# I will create a new feature Children, from merging Kidhome an Teenhome
# I will create the feature 'Total_Offers'that indicate the number of offers a customer accepted all offer period.
# I will create the feature 'Total_Spent' which is the total amount spent on the the household goods features: MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds 
# Next, I will be working on the categorical features
```

```python
# Creating the Age feature
data['Age'] = 2022 - data['Year_Birth']

# Converting to Datetime object and checking for the most recent record
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer']).apply(lambda x: x.date())
latest_customer = data['Dt_Customer'].max()

# Creating the join_duration feature
data['Join_Duration'] = latest_customer - data['Dt_Customer'] 
data['Join_Duration'] = pd.to_numeric(data['Join_Duration'], errors='coerce')

# Creating the feature Children
data['Children'] = data['Kidhome'] + data['Teenhome']

# Creating the Total_Offers feature
data['Total_Offers'] = data[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)

# Creating the Total_Spent feature
data['Total_Spent'] = data[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# I will eqaully be engineering the categorical features to make them more arranged and concise

# Reassigning the categories of the Education feature
data['Education'] = data['Education'].replace({'Graduation': 'Graduate', 'PhD': 'Postgraduate', 'Master':'Postgraduate', '2n Cycle':'Postgraduate', 'Basic': 'Undergraduate'})

# Reassigning the categories of the Marital_Status feature
data['Marital_Status'] = data['Marital_Status'].replace({'Married': 'Couple', 'Together': 'Couple', 'Single': 'Non_Couple', 'Divorced': 'Non_Couple', 'Widow': 'Non_Couple', 'Alone':'Non_Couple', 
                                                         'YOLO':'Non_Couple', 'Absurd':'Non_Couple'})

# Will go ahead and drop redundant features
data = data.drop(['Dt_Customer','Year_Birth'],axis=1)
```

# Data Preprocessing

```python
# In this section, I will check out for the categorical features in the data and convert them into a numeric form, scale the features and ready them for clustering
```

```python
data.select_dtypes('object').columns
```

```python
# Converting categorical features to numeric
df = data.copy()
df = pd.get_dummies(data, drop_first=True)
```

```python
# Here, I will be regularizing all the features to be on the same scale.I however will not be involving features suc as AcceptedCmp1 and other 'numerical boolean' features, as these areb technoically not numeric.
# They are discrete, since they represent no/yes. Scaling them will lead to the losss of their original interpretation.

#to_scale = [col for col in data.select_dtypes('int64').columns if len(data[col].value_counts())> 2]
to_scale = ['Income','Kidhome','Teenhome','Recency','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases','NumWebPurchases',
            'NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth','Age','Join_Duration','Children','Total_Spent']
scaler = StandardScaler()

for scaled in df[to_scale].columns:
    df[scaled] = (scaler.fit_transform(df[[scaled]]))
df.head()
```

```python
# Factor Analysis

# Now,just as mentioned ealier, it is almost often the case that some features are related to one another, thus, their presence would be unnecessarily redunt. Since including too many features in the model
# could potentially render our model less performant, I will be performing factor analysis. Factor analysis is a statistical technique used to examine the interrelationships in a set og features, to see the
# features with the most variability
```

```python
pca = PCA(n_components=2)
pca.fit(df)
data_decomp = pca.transform(df)
df_decomp = pd.DataFrame(data_decomp, columns = ['first_col', 'second_col'])
df_decomp.head()
```

```python
# We can visualize the new two principal components in a scatter plot

sns.jointplot(data=df_decomp, x=df_decomp['first_col'], y=df_decomp['second_col'])
```

# CLUSTERING

```python
# In this section, I will be performing clustering on the two principal features to segment the customers into groups, where all the customers in each group have common purchasing habits. 
# This will be done using K MEANS.

# Generally, when working in clustering, there is no wrong or right answer to the number of groups that we should segment our data into.The common approach to have an idea is to use the Elbow Method.
# Hence, I will be usin the Elbow Method to select the optimum number of groups, and then visualize the groups aferwards.
```

```python
distortions = []
K = range(1,11)

for k in K:
    k_model = KMeans(n_clusters=k)
    k_model.fit(df_decomp)
    distortions.append(k_model.inertia_)


plt.figure(figsize=(15,9))
plt.plot(K,distortions)
plt.scatter(K,distortions)
plt.xlabel('K')
plt.ylabel('Distortions')
plt.title('Elbow Method for Choosing the Number of Groups')
```

```python
# As shown by the Elbow Method, the optimal cluster number to choose is where k=4. As a result, I will perform the clustering by segmenting the customers into four groups.
```

```python
actual_model = KMeans(n_clusters=4)
actual_model.fit(df_decomp)
k_predict = actual_model.predict(df_decomp)

# Since we have the model grouping the customers into the 4 clusters now, we can add the cluster to our data set for further visualization
data['Group'] = k_predict
```
